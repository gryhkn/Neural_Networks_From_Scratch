# Neural Networks with One Layer

### Single NN with input

```python
inputs1 = [1, 2, 3]
weights1 = [0.2, 0.8, -0.5]
bias1 = 2

output1 = (inputs1[0]*weights1[0] +
          inputs1[1]*weights1[1] +
          inputs1[2]*weights1[2] + bias1)

print(output1)
```

###  Single NN with 4 inputs

```python
inputs2 = [1.0, 2.0, 3.0, 2.5]
weights2 = [0.2, 0.8, -0.5, 1.0]
bias2 = 2.0

output2 = (inputs2[0]*weights2[0] +
          inputs2[1]*weights2[1] +
          inputs2[2]*weights2[2] +
          inputs2[3]*weights2[3] + bias2)

print(output2)
```

### A Layer of NN's with 4 inputs

```python
inputs = [1, 2, 3, 2.5]

weights1 = [0.2, 0.8, -0.5, 1]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]

bias1 = 2
bias2 = 3
bias3 = 0.5

outputs = [
    #Neuron 1
    inputs[0]*weights1[0] + inputs[1]*weights1[1] +
    inputs[2]*weights1[2] + inputs[3]*weights1[3] + bias1,
    
    #Neuron 2
    inputs[0]*weights2[0] + inputs[1]*weights2[1] +
    inputs[2]*weights2[2] + inputs[3]*weights2[3] + bias2,
    
    #Neuron 3
    inputs[0]*weights3[0] + inputs[1]*weights3[1] +
    inputs[2]*weights3[2] + inputs[3]*weights3[3] + bias3
]

print(outputs)
```

In this code, we have three sets of weights and three biases, which define three neurons. Each neuron is “connected” to the same inputs. The difference is in the separate weights and biasthat each neuron applies to the input. This is called a **fully connected** neural network — every neuron in the current layer has connections to every neuron from the previous layer. 

### Some optimizations in code

```python
inputs = [1, 2, 3, 2.5]
weights = [[0.2, 0.8, -0.5, 1],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

# Output of current layer
layer_outputs = []

# for each neuron
for neuron_weights, neuron_bias in zip(weights, biases):
    print(f"Neuron weights: {neuron_weights} \
    Neuron bias: {neuron_bias}")
    
    #Zeroed output of given neuron
    neuron_output = 0
    
    # For each input and weight to the neuron
    for n_input, weight in zip(inputs, neuron_weights):
        # Multiply this input by assosiated weight
        # and add to the neuron's output variable
        neuron_output += n_input*weight
    
    # Add bias
    neuron_output += neuron_bias
    
    # Put neuron's result to the layer's output list
    layer_outputs.append(neuron_output)

print(f"layer outputs: {layer_outputs}")
```

* This is a solution but when we have more than 1D inputs, like matrix or 3D, using for loop would be useless. Instead, we can use Numpy **.dot** method. Using .dot method is below.

## 3 Neurons:

```python
import numpy as np

inputs = [[1.0, 2.0, 3.0, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, -0.8]]

# Our input is not a list and weights as well.
# They're matrix's. So we have to make matrix product.

weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2.0, 3.0, 0.5]

layer_outputs = np.dot(inputs, np.array(weights).T) + biases

print(f"Layer outputs: \n {layer_outputs}")
```
