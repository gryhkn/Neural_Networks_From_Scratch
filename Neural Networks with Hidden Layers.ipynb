* Neural networks become **“deep”** when they have 2 or more **hidden layers**.

* A hidden layer is not an input or output layer. We (as machine learning engineers/data scientist) can see just inputs and outputs. Layers between these endpoints( inputs --like image, .csv, .txt, videos etc-- and outputs) have values that we don't necesserily dealt with. So this is why we call them **hidden**. 

Samples (feature set data) get fed through the input, which does not change it in any way, to ourfirst hidden layer, which we can see has 3 sets of weights, with 4 values each.

Each of those 3 unique weight sets is associated with its distinct neuron. Thus, since we have 3 weight sets, we have 3 neurons in this first hidden layer. Each neuron has a unique set of weights,of which we have 4 (as there are 4 inputs to this layer), which is why our initial weights have ashape of (3,4).

---

## NN w/ Hidden Layer

```python
import numpy as np

inputs = [[1.0, 2.0, 3.0, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, -0.8]]

weights1 = [[0.2, 0.8, -0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5],
            [-0.26, -0.27, 0.17, 0.87]]

biases1 = [2.0, 3.0, 0.5]

weights2 = [[0.1, -0.14, 0.5],
           [-0.5, 0.12, -0.33],
           [-0.44, 0.73, -0.13]]

biases2 = [-1, 2, -0.5]

layer1_output = np.dot(inputs, np.array(weights1).T) + biases1

layer2_output = np.dot(layer1_output, np.array(weights2).T) + biases2


print(layer2_output)
```
